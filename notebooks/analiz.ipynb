```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri yükle
df = pd.read_csv('../data/ornek-veri.csv')

# İlk 5 satırı göster
print(df.head())

# Temel istatistikler
print(df.describe())

# Yaş ve maaş ilişkisi
plt.figure(figsize=(18,5))
sns.scatterplot(data=df, x='yas', y='maas')
plt.title('Yaş vs Maaş')
plt.show()
```

```python
df.groupby("maas")["yas"].agg(["mean", "median", "count"])
```

```python
sns.countplot(x="yas", data=df)
```

```python
sns.heatmap(df[['yas', 'maas']].corr(), annot=True, cmap="coolwarm")
```
